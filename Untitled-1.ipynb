```python
import pandas as pd
import numpy as np
from torch import nn
from sklearn.preprocessing import StandardScaler

train_data = pd.read_excel(r'E:\唐筛\train.csv')
test_data = pd.read_csv(r'E:\唐筛\test.csv')
diag_1_data = train_data[train_data['诊断'] == 1]
diag_1_data = diag_1_data.sample(n=10000, random_state=42)
diag_2_data = train_data[train_data['诊断'] == 2]
merge_data = pd.concat([diag_1_data,diag_2_data], ignore_index=True)
merge_data = merge_data.sample(frac=1, random_state=42).reset_index(drop=True)

scaler = StandardScaler()
merge_data[['AFPMOM','BHCGMOM','UE3MOM']] = scaler.fit_transform(merge_data[['AFPMOM',
                                                                'BHCGMOM','UE3MOM']])
test_data[['AFPMOM','BHCGMOM','UE3MOM']] = scaler.transform(test_data[['AFPMOM',
                                                                'BHCGMOM','UE3MOM']])
merge_data[['体重', '年龄']] = scaler.fit_transform(merge_data[['体重','年龄']])
test_data[['体重', '年龄']] = scaler.transform(test_data[['体重','年龄']])
```

```python
X_train = merge_data.iloc[:, [1, 2, 4, 5, 6]].values
y_train = merge_data.iloc[:, 0].values
X_test = test_data.iloc[:, [1, 2, 3, 4, 5]].values
y_test = test_data.iloc[:, -1].values
```

```python
y_train[y_train == 1] = 0
y_train[y_train == 2] = 1
y_test[y_test == 1] = 0
y_test[y_test == 2] = 1
```

```python
import torch
from torch.utils.data import TensorDataset, DataLoader 
from d2l import torch as d2l

num_inputs, num_hidden, num_outputs = 5, 5, 1

net = nn.Sequential(nn.Linear(num_inputs, num_hidden),
                    nn.ReLU(),
                    nn.Dropout(0.5),
                    nn.Linear(num_hidden, num_outputs),
                    nn.Sigmoid())

X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)  
y_test_tensor = torch.tensor(y_test, dtype=torch.float32).unsqueeze(1)

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

batch_size = 64
train_iter = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_iter = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

lr = 0.0001
num_epochs = 50
loss = nn.BCELoss()
trainer = torch.optim.Adam(net.parameters(), lr=lr)
```

```python
import matplotlib.pyplot as plt

def init_weights(m):
    if type(m) == nn.Linear:
        nn.init.kaiming_uniform_(m.weight)

net.apply(init_weights) 

prev_test_acc = 0.0
count = 0

train_losses = []
train_accuracies = []
test_accuracies = []

lambda_ = 0.001
def l1_regularization(model, lambda_):
    l1_reg = 0.0
    for param in model.parameters():
        l1_reg += lambda_ * torch.norm(param, 1)
    return l1_reg

# 训练循环
for epoch in range(num_epochs):
    net.train()
    train_l_sum, train_acc_sum, num_examples = 0.0, 0.0, 0
    for X, y in train_iter:
        y_hat = net(X)
        y = y.float()
        l = loss(y_hat, y) + l1_regularization(net, lambda_)
        trainer.zero_grad()
        l.backward()
        trainer.step()
        train_l_sum += l.item()
        train_acc_sum += ((y_hat > 0.5).type(torch.float32) == y).sum().item()
        num_examples += y.shape[0]

    # 计算测试集准确度
    net.eval()
    test_acc = 0.0
    num_test_examples = 0
    with torch.no_grad():
        for X_test, y_test in test_iter:
            y_test_hat = net(X_test)
            y_test = y_test.float()
            test_acc += ((y_test_hat > 0.5).type(torch.float32) == y_test).sum().item()
            num_test_examples += y_test.shape[0]

    test_acc /= num_test_examples
    
    train_losses.append(train_l_sum / num_examples)
    train_accuracies.append(train_acc_sum / num_examples)
    test_accuracies.append(test_acc)
    # 打印训练和测试结果
    print(f'Epoch {epoch + 1}, Train loss: {train_l_sum / num_examples}, '
          f'Train accuracy: {train_acc_sum / num_examples}, '
          f'Test accuracy: {test_acc}')

    prev_test_acc = test_acc
    
model_params = net.state_dict()

for param_name, param_value in model_params.items():
    print(f'{param_name}: {param_value}')
    
plt.plot(train_accuracies, label='Train Accuracy')
plt.plot(test_accuracies, label='Test Accuracy')
plt.plot(train_losses, label='Train Loss')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Train Loss, Train and Test Accuracy during Training')
plt.legend()
plt.ylim(0.0, 1.0)
plt.tight_layout()
plt.show()
```
